### Data Preprocessing

```python
data=data.drop('20-day MA',axis=1)
```

```python
data['date'] = pd.to_datetime(data['date'])
stock_data = data.sort_values('date')
```

```python
stock = data[['date', 'close', 'high', 'low', 'open', 'volume']]
```

### Data Normalization

```python
from sklearn.preprocessing import MinMaxScaler
from sklearn.model_selection import train_test_split
```

```python
scaler=MinMaxScaler(feature_range=(0,1))
```

```python
normalize_data = stock[['open', 'high', 'low', 'volume', 'close']].copy()
normalize_data = scaler.fit_transform(normalize_data)
```

```python
train_data, test_data = train_test_split(normalize_data, train_size=0.8, test_size=0.2, shuffle=False)
```

```python
train = pd.DataFrame(train_data, columns=['open', 'high', 'low', 'volume', 'close'])
test = pd.DataFrame(test_data, columns=['open', 'high', 'low', 'volume', 'close'])
```

```python

#Sequence Generation

def generate_sequences(df, seq_length=50):
    X = df[['open', 'high', 'low', 'volume', 'close']].reset_index(drop=True)
    y = df[['open', 'high', 'low', 'volume', 'close']].reset_index(drop=True)
    sequences = []
    labels = []
    #iterate over the data to create sequences
    for index in range(len(X) - seq_length + 1):
        sequences.append(X.iloc[index : index + seq_length].values)
        labels.append(y.iloc[index + seq_length - 1].values)
    #convert the list to numpy arrays
    sequences = np.array(sequences)
    labels = np.array(labels)

    return sequences, labels
```

```python
train_seq, train_labels = generate_sequences(train)
test_seq, test_labels = generate_sequences(test)
```

### Model Architecture

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense, Dropout
```

```python

#Model Building

model = Sequential()
model.add(LSTM(units=50, return_sequences=True, input_shape=(50, 5))),
model.add(Dropout(0.2)),
model.add(LSTM(units=50, return_sequences=True)),
model.add(Dropout(0.2)),
model.add(LSTM(units=50)),
model.add(Dropout(0.2)),
model.add(Dense(units=5))
```

```python
model.summary()
```

```python
model.compile(optimizer='adam', loss='mean_squared_error', metrics=['mean_squared_error'])
```

```python
#train the model
history = model.fit(train_seq,train_labels,epochs=200,batch_size=32,validation_data=(test_seq, test_labels),verbose=1)
```

```python
history.history
```

```python
train_predictions = model.predict(train_seq)
test_predictions = model.predict(test_seq)
```

### Visualize the predictions

```python
from plotly.subplots import make_subplots
```

```python

#plot for close predictions

fig = make_subplots(rows=1, cols=1, subplot_titles=('Close Predictions'))
train_close_pred = train_predictions[:, 0]
train_close_actual = train_labels[:, 0]
fig.add_trace(go.Scatter(x=np.arange(len(train_close_actual)), y=train_close_actual, mode='lines', name='Actual', opacity=0.9))
fig.add_trace(go.Scatter(x=np.arange(len(train_close_pred)), y=train_close_pred, mode='lines', name='Predicted', opacity=0.6))
fig.update_layout(title='Close Predictions', template='plotly_dark')
fig.show()
```

```python

#next 10 days latest predictions

latest_predict = []
last_seq = test_seq[:-1]
for _ in range(10):
    prediction = model.predict(last_seq)
    latest_predict.append(prediction)
```

```python

#plot for next 10 days predictions

pi.templates.default = "plotly_dark"
predict_next = np.array(latest_predict).reshape(-1, 5)
last_date = stock['date'].max()
next_10_days = [last_date + timedelta(days=i) for i in range(1, 11)]
for i, feature_name in enumerate(['open', 'high', 'low', 'volume', 'close']):
    if feature_name in ['volume', 'close']:
        fig = go.Figure()
        fig.add_trace(go.Scatter(x=next_10_days, y=predict_next[:, i],
                                 mode='lines+markers', name=f'Predicted {feature_name.capitalize()} Prices'))

        fig.update_layout(title=f'Predicted {feature_name.capitalize()} Prices for the Next 10 Days',
                          xaxis_title='Date', yaxis_title=f'{feature_name.capitalize()} Price')
        fig.show()
```
